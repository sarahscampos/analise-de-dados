- **Distribuição amostral**: Refere-se à distribuição das estatísticas calculadas a partir de múltiplas amostras retiradas de uma população. Por exemplo, se coletamos várias amostras e calculamos a média de cada uma, a distribuição dessas médias segue um padrão específico, frequentemente aproximando-se de uma distribuição normal, conforme o Teorema Central do Limite.

- **Intervalo de confiança**: É um intervalo de valores dentro do qual se espera que um parâmetro populacional verdadeiro esteja, com determinada probabilidade (chamada nível de confiança). Por exemplo, um intervalo de confiança de 95% indica que, se repetirmos a amostragem várias vezes, em 95% dos casos o valor verdadeiro do parâmetro estará dentro do intervalo calculado.

# Featuring engineering
Processo de criar novas variáveis a partir dos dados existentes para melhorar a análise e a modelagem

```python
def preencher_nivel(gestor, nivel):
  if gestor==1:
    return "Gestora"
  else:
    return nivel
```

```python
dados['NOVO_NIVEL'] = dados.apply(lambda x: preencher_nivel(x['GESTOR?'], x['NIVEL']), axis=1)
```

```python
dados['NOVO_NIVEL'].value_counts()
```

```python
dados = pd.get_dummies(dados, columns=['NIVEL']) #transformando dados categoricos em booleanos
```

```python
dados.columns
```

```python
def determinar_geracao(idade):
  if 39<idade<=58:
    return "Geração X"
  elif 29<idade<=39:
    return "Millenial"
  elif 13<idade<=29:
    return "Geração Z"
  else:
    return "Outra geração"
```

```python
dados['GERACAO'] = dados['IDADE'].apply(determinar_geracao)
```

```python
dados['GERACAO'].value_counts()
```
