# Visualização de dados

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
dados = pd.read_csv('/content/drive/MyDrive/Programaria/analise_dados2.csv')
```

```python
dados.head()
```

```python
genero_qtd = dados['GENERO'].value_counts()
```

```python
plt.figure()
plt.bar(height = genero_qtd.values, x = genero_qtd.index)
plt.title('Quantidade de pessoas por gênero')
plt.xlabel('Gênero')
plt.ylabel('Contagem')
plt.show()
```

```python
plt.figure()
sns.countplot(data=dados, x='GENERO', palette='pastel')
plt.title('Quantidade de pessoas por gênero')
plt.xlabel('Gênero')
plt.ylabel('Contagem')
plt.show()
```

```python
salario_idade = dados.groupby("IDADE")["SALARIO"].mean()
```

```python
salario_idade
```

```python
plt.figure()
plt.plot(salario_idade.index, salario_idade.values, marker='o', linestyle='--')
plt.xlabel('Idade')
plt.ylabel('Média de salário')
plt.title('Média de salário por idade')
plt.grid(True)
plt.show()
```

```python
fig = px.line(salario_idade.reset_index(), x='IDADE', y='SALARIO', title='Média de salário por idade', markers=True)
fig.show()
```

```python
plt.figure(figsize=(15,5))
plt.scatter(dados['IDADE'], dados['SALARIO'], alpha=0.5)
plt.xlabel('Idade')
plt.ylabel('Salário')
plt.title('Dispersão de salário por idade')
plt.grid(True)
plt.show()
```

```python
fig = px.scatter(dados, x='IDADE', y='SALARIO', title='Dispersão de salário por idade')
fig.show()
```
